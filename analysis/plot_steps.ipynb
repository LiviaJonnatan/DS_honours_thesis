# Explore Bovine Steps

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
pd.set_option("mode.chained_assignment", None)
```

```python
gen_df = pd.read_csv("gen_mean_step_diff.csv")
gen_df.info()
```

```python
estrus_df = gen_df[gen_df.estrus==1]
nonestrus_df = gen_df[gen_df.estrus==0]
```

## Plot: overall hourly mean step difference

### 30 minutes interval

```python
diff_means_overall = gen_df[["hour","step_diff_30min_mean_overall"]].drop_duplicates().sort_values(by=["hour"], ascending = True)
```

```python
temp = estrus_df.groupby(["hour"]).agg(step_diff_30min_mean_overall = ("step_diff_30min", "mean"))
temp = temp.reset_index()

hr = temp.hour.tolist()
diff_means_overall_est = temp.step_diff_30min_mean_overall.tolist()

# Fill calculated mean into original dataframe
estrus_df["step_diff_30min_mean_overall"] = 0
for i in range(len(hr)):
    estrus_df.step_diff_30min_mean_overall[(estrus_df.hour == hr[i])] = diff_means_overall_est[i]
```

```python
temp = nonestrus_df.groupby(["hour"]).agg(step_diff_30min_mean_overall = ("step_diff_30min", "mean"))
temp = temp.reset_index()

hr = temp.hour.tolist()
diff_means_overall_nonest = temp.step_diff_30min_mean_overall.tolist()

# Fill calculated mean into original dataframe
nonestrus_df["step_diff_30min_mean_overall"] = 0
for i in range(len(hr)):
    nonestrus_df.step_diff_30min_mean_overall[(nonestrus_df.hour == hr[i])] = diff_means_overall_nonest[i]
```

```python
plt.figure(figsize=(15, 12))
plt.scatter(hr, diff_means_overall_nonest, c="b", label="Avg non estrus")
plt.scatter(hr, diff_means_overall_est, c="r", label="Avg estrus")
plt.plot(hr, diff_means_overall["step_diff_30min_mean_overall"], "--g",label="Avg overall")
plt.xticks(hr, hr)
plt.legend(loc="upper left")
plt.xlabel("Hour")
plt.ylabel("Step difference")
plt.title("Mean of step difference (30 min.) for all cows by hour")
plt.show()
```

## Plot: hourly mean step difference by cow

### 30 minutes interval

```python
diff_means_overall = gen_df[["hour","step_diff_30min_mean_overall"]].drop_duplicates().sort_values(by=["hour"], ascending = True)
```

```python
temp = estrus_df.groupby(["AnimalID", "hour"]).agg(step_diff_30min_mean_by_cow = ("step_diff_30min", "mean"))
temp = temp.reset_index()

ids = temp["AnimalID"].tolist()
hr_by_cow_est = temp.hour.tolist()
diff_means_by_cow_est = temp.step_diff_30min_mean_by_cow.tolist()

# Fill calculated mean into original dataframe
estrus_df["step_diff_30min_mean_by_cow"] = 0
for i in range(len(ids)):
    estrus_df.step_diff_30min_mean_by_cow[(estrus_df.AnimalID == ids[i]) & (estrus_df.hour == hr_by_cow_est[i])] = diff_means_by_cow_est[i]
```

```python
plt.figure(figsize=(15, 12))
plt.plot(hr, diff_means_overall_nonest, "--b", label="Avg non estrus")
plt.plot(hr, diff_means_overall_est, "--r", label="Avg estrus")
plt.plot(hr, diff_means_overall["step_diff_30min_mean_overall"], "--g",label="Avg overall")
plt.scatter(hr_by_cow, diff_means_by_cow_nonest, c="y", label="Non estrus")
plt.scatter(hr_by_cow_est, diff_means_by_cow_est, c="tab:pink", label="Estrus")
plt.xticks(hr, hr)
plt.legend(loc="upper left")
plt.xlabel("Hour")
plt.ylabel("Step difference")
plt.title("Mean of step difference (30 min.) by cow, by hour")
plt.show()
```
